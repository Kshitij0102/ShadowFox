```python
from transformers import pipeline

generator = pipeline("text-generation", model="gpt2")
```

```python
prompt = "Artificial Intelligence will change the world because"

result = generator(prompt, max_length=60, num_return_sequences=3)

for r in result:
    print(r['generated_text'])
    print()
```

```python
prompts = [
    "The future of education will",
    "Machine learning helps businesses",
    "The biggest challenge in AI is"
]

for p in prompts:
    print("PROMPT:", p)
    out = generator(p, max_length=50)[0]['generated_text']
    print("OUTPUT:", out)
    print("-"*50)
```
